# Feature Engineering 

```python
df.columns
```

```python
# 1. Sales (Revenue per transaction)
df['Sales'] = df['Quantity'] * df['UnitPrice'] * (1 - df['Discount']/100)
```

```python
# 2. Revenue
df['Revenue'] = df['Quantity'] * df['UnitPrice'] * (1 - df['Discount']/100)
```

```python
# 3. Return Flag (1 = Returned, 0 = Not Returned)
df['IsReturn'] = df['Quantity'].apply(lambda x: 1 if x < 0 else 0)
```

```python
# 4. Order Size Category
df['OrderSize'] = pd.cut(df['Quantity'], 
                         bins=[-100,0,5,20,100,1000],
                         labels=['Return','Small','Medium','Large','Bulk'])
```

```python
# 5. Net Revenue after shipping
df['NetRevenue'] = df['Revenue'] - df['ShippingCost']
```

```python
# 6. Customer Lifetime Value (total sales per customer)
df['CustomerLTV']  = df.groupby('CustomerID')['Revenue'].transform('sum')
```

```python
# 8. Discount Flag (high discount vs low discount)
df['HighDiscount'] = df['Discount'].apply(lambda x: 1 if x > 30 else 0)
```

```python
df['Year'] = df['InvoiceDate'].dt.year
df['MonthName'] = df['InvoiceDate'].dt.month_name()
df['Day'] = df['InvoiceDate'].dt.day
df['Weekday'] = df['InvoiceDate'].dt.day_name()
df['Hour'] = df['InvoiceDate'].dt.hour
```

```python
df.head()
```

```python
df[['CustomerLTV']]## outlier detect
```

```python
# Step 1: Calculate LTV per customer
customer_ltv = df.groupby('CustomerID')['Revenue'].sum().reset_index()

# Step 2: Use IQR method to remove outliers
Q1 = customer_ltv['Revenue'].quantile(0.25)
Q3 = customer_ltv['Revenue'].quantile(0.75)
IQR = Q3 - Q1

upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

# Step 3: Filter out outliers
customer_ltv_clean = customer_ltv[
    (customer_ltv['Revenue'] >= lower_bound) & 
    (customer_ltv['Revenue'] <= upper_bound)
]
```

```python
outlier_customers = customer_ltv[
    (customer_ltv['Revenue'] < lower_bound) | 
    (customer_ltv['Revenue'] > upper_bound)
]['CustomerID']

# Step 4: Create CustomerLTV column without outliers
df['CustomerLTV'] = df.groupby('CustomerID')['Revenue'].transform('sum')

# Step 5: Remove outliers from that column
df.loc[df['CustomerID'].isin(outlier_customers), 'CustomerLTV'] = np.nan
median_ltv = df['CustomerLTV'].median()
df['CustomerLTV'] = df['CustomerLTV'].fillna(median_ltv)
```
